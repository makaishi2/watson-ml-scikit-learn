# CIFAR-10 Webアプリのテスト用パラメータを作る

```python
# 必要ライブラリ
from keras.datasets import cifar10
import matplotlib.pyplot as plt
from PIL import Image
from scipy.misc import toimage
import numpy as np
import json
```

```python
# ラベル定義
labels = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
# イメージデータ読み込み
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# Indexのシャフル
indexes = list(range(len(y_test)))
np.random.shuffle(indexes)
indexes_sub = indexes[:10]
```

```python
print(indexes_sub)
```

```python
# 表示関数定義
# Index番目のテストデータの画面表示と正解データ
def disp_image(index):
    a_ind = y_test[index].flat[0]
    answer = labels[a_ind]
    title = 'index:%d  %s  (%d)' % (index, answer, a_ind)
    img = Image.fromarray(x_test[index])
    plt.figure(figsize=(2,2))
    plt.imshow(img)
    plt.axis('off')
    plt.title(title, fontsize=16)
    plt.show()
```

```python
# Deep Learing WEBサービス
# パラメータ作成関数
#
def get_param(index):
    x_param1 = x_test[index].astype("float32") / 255.0
    x_param2 = x_param1.reshape( (1,) + x_param1.shape).tolist()
    # print(x_param2.shape)
    # print(x_param2)
    params = {}
    params['values'] = x_param2
    return params    
```

```python
for index in indexes_sub:
    disp_image(index)
    print(json.dumps(get_param(index)))
```
